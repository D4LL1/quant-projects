```python
import pandas as pd
import requests
import gzip
import io
import time
from concurrent.futures import ThreadPoolExecutor, as_completed
from typing import List, Dict, Optional
import warnings
warnings.filterwarnings('ignore')

class OptimizedBybitFetcher:
    def __init__(self, max_workers=15, request_timeout=20, verbose=False):
        """
        Zoptymalizowany fetcher z minimalnymi logami i maksymalną wydajnością.
        
        Args:
            max_workers: Maksymalna liczba równoczesnych wątków
            request_timeout: Timeout dla pojedynczego żądania
            verbose: Czy pokazywać szczegółowe informacje
        """
        self.max_workers = max_workers
        self.request_timeout = request_timeout
        self.verbose = verbose
        self.headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
            'Accept-Encoding': 'gzip, deflate',
            'Connection': 'keep-alive'
        }
        
        # Session dla connection pooling
        self.session = requests.Session()
        self.session.headers.update(self.headers)
    
    def _fetch_single_date(self, ticker: str, date: str) -> Optional[pd.DataFrame]:
        """
        Pobiera dane dla pojedynczej daty - zoptymalizowane.
        """
        url = f'https://public.bybit.com/spot/{ticker}/{ticker}_{date}.csv.gz'
        
        try:
            response = self.session.get(url, timeout=self.request_timeout)
            response.raise_for_status()
            
            # Szybka dekompresja i wczytanie
            with gzip.open(io.BytesIO(response.content), 'rt') as f:
                df = pd.read_csv(f)
            
            if not df.empty:
                df['ticker'] = ticker
                
                # Ustaw index na timestamp
                timestamp_col = 'timestamp' if 'timestamp' in df.columns else 'time'
                if timestamp_col in df.columns:
                    df = df.set_index(timestamp_col)
                
                return df
                
        except requests.exceptions.HTTPError as e:
            if self.verbose and e.response.status_code != 404:
                print(f"HTTP error {ticker} {date}: {e.response.status_code}")
        except Exception as e:
            if self.verbose:
                print(f"Error {ticker} {date}: {str(e)[:50]}")
        
        return None
    
    def fetch_ticker(self, ticker: str, start_date: str, end_date: str) -> pd.DataFrame:
        """
        Pobiera dane dla jednego tickera z zakresu dat - wielowątkowe.
        """
        date_range = pd.date_range(start_date, end_date)
        date_list = date_range.strftime('%Y-%m-%d').tolist()
        
        if self.verbose:
            print(f"Fetching {ticker}: {len(date_list)} dates")
        
        dfs = []
        successful_dates = 0
        
        # Wielowątkowe pobieranie
        with ThreadPoolExecutor(max_workers=self.max_workers) as executor:
            future_to_date = {
                executor.submit(self._fetch_single_date, ticker, date): date 
                for date in date_list
            }
            
            for future in as_completed(future_to_date):
                df = future.result()
                if df is not None and not df.empty:
                    dfs.append(df)
                    successful_dates += 1
        
        # Kombinuj wyniki
        if dfs:
            combined_df = pd.concat(dfs, ignore_index=False, sort=True)
            combined_df = combined_df.sort_index()
            
            if self.verbose:
                print(f"✓ {ticker}: {len(combined_df):,} records ({successful_dates}/{len(date_list)} days)")
            
            return combined_df
        else:
            if self.verbose:
                print(f"✗ {ticker}: No data found")
            return pd.DataFrame()
    
    def fetch_multiple_tickers(self, tickers: List[str], start_date: str, 
                             end_date: str, combine_data: bool = True) -> Dict:
        """
        Pobiera dane dla wielu tickerów - zoptymalizowane wielowątkowo.
        """
        start_time = time.time()
        results = {}
        
        print(f"Fetching data for {len(tickers)} tickers ({start_date} to {end_date})")
        
        # Przetwarzaj wszystkie tickery równocześnie
        with ThreadPoolExecutor(max_workers=min(len(tickers), 5)) as executor:
            future_to_ticker = {
                executor.submit(self.fetch_ticker, ticker, start_date, end_date): ticker 
                for ticker in tickers
            }
            
            for i, future in enumerate(as_completed(future_to_ticker), 1):
                ticker = future_to_ticker[future]
                try:
                    df = future.result()
                    if not df.empty:
                        results[ticker] = df
                        print(f"[{i}/{len(tickers)}] ✓ {ticker}: {len(df):,} records")
                    else:
                        print(f"[{i}/{len(tickers)}] ✗ {ticker}: No data")
                except Exception as e:
                    print(f"[{i}/{len(tickers)}] ✗ {ticker}: Error - {str(e)[:50]}")
        
        elapsed_time = time.time() - start_time
        total_records = sum(len(df) for df in results.values())
        
        print(f"\nCompleted: {len(results)}/{len(tickers)} tickers, {total_records:,} total records")
        print(f"Time: {elapsed_time:.1f}s ({total_records/elapsed_time:.0f} records/sec)")
        
        if not results:
            return pd.DataFrame() if combine_data else {}
        
        if combine_data:
            combined_df = pd.concat(results.values(), ignore_index=False, sort=True)
            combined_df = combined_df.sort_index()
            print(f"Combined DataFrame: {len(combined_df):,} records")
            return combined_df
        else:
            return results
    
    def __del__(self):
        """Zamknij session przy usuwaniu obiektu."""
        if hasattr(self, 'session'):
            self.session.close()

def fetch_bybit_data(tickers: List[str], start_date: str, end_date: str, 
                    max_workers: int = 15, combine_data: bool = True, 
                    verbose: bool = False):
    """
    Główna funkcja do pobierania danych z Bybit.
    
    Args:
        tickers: Lista symboli do pobrania
        start_date: Data początkowa (YYYY-MM-DD)
        end_date: Data końcowa (YYYY-MM-DD)
        max_workers: Liczba równoczesnych wątków
        combine_data: Czy połączyć dane w jeden DataFrame
        verbose: Czy pokazywać szczegółowe logi
    
    Returns:
        DataFrame lub Dict z DataFrames
    """
    fetcher = OptimizedBybitFetcher(max_workers=max_workers, verbose=verbose)
    return fetcher.fetch_multiple_tickers(tickers, start_date, end_date, combine_data)


def fetch_single_ticker(ticker: str, start_date: str, end_date: str,
                       max_workers: int = 15, verbose: bool = False):
    """
    Pobiera dane dla pojedynczego tickera.
    """
    fetcher = OptimizedBybitFetcher(max_workers=max_workers, verbose=verbose)
    return fetcher.fetch_ticker(ticker, start_date, end_date)


def quick_test(ticker: str, date: str):
    """
    Szybki test pojedynczej daty.
    """
    fetcher = OptimizedBybitFetcher(verbose=True)
    result = fetcher._fetch_single_date(ticker, date)
    
    if result is not None:
        print(f"✓ Success: {len(result)} records")
        print(f"Columns: {list(result.columns)}")
        print(result.head())
        return result
    else:
        print("✗ No data or error")
        return None

    pass
```

```python
from typing import Dict
import gc

def process_ticker_data(df: pd.DataFrame) -> Dict[str, pd.DataFrame]:
    """
    Funkcja do przetwarzania danych tickerów z optymalizacją pamięci
    Timestamp jest w indeksie, kolumny: price, ticker, volume
    
    Parameters:
    df (pd.DataFrame): DataFrame z indeksem timestamp i kolumnami: price, ticker, volume
    
    Returns:
    Dict[str, pd.DataFrame]: Słownik z DataFrames dla każdego tickera
    """
    
    print(f"Przetwarzanie {len(df)} rekordów...")
    print(f"Kolumny: {list(df.columns)}")
    print(f"Indeks: {df.index.name}")
    
    # Kopiowanie DataFrame żeby nie modyfikować oryginału
    df_work = df.copy()
    
    # Optymalizacja typów danych dla oszczędności pamięci
    print("Optymalizacja typów danych...")
    df_work['price'] = pd.to_numeric(df_work['price'], downcast='float')
    df_work['volume'] = pd.to_numeric(df_work['volume'], downcast='float')
    df_work['ticker'] = df_work['ticker'].astype('category')  # Categoria dla powtarzających się stringów
    
    # Konwersja timestamp z indeksu na datetime (z milisekund)
    print("Konwersja timestampów...")
    df_work['datetime'] = pd.to_datetime(df_work.index, unit='ms')
    
    # Utworzenie kolumny z czasem do sekundy (bez milisekund)
    df_work['time_key'] = df_work['datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
    
    # Słownik do przechowywania DataFrames dla każdego tickera
    ticker_dataframes = {}
    
    # Pobieranie unikalnych tickerów
    unique_tickers = df_work['ticker'].unique()
    print(f"Znaleziono {len(unique_tickers)} unikalnych tickerów: {list(unique_tickers)}")
    
    # Przetwarzanie każdego tickera osobno
    for i, ticker in enumerate(unique_tickers):
        print(f"Przetwarzanie tickera {i+1}/{len(unique_tickers)}: {ticker}")
        
        # Filtrowanie danych dla konkretnego tickera
        ticker_data = df_work[df_work['ticker'] == ticker].copy()
        
        # Ustawianie indeksu na time_key
        ticker_data = ticker_data.set_index('time_key')
        
        # Grupowanie po sekundach i sumowanie
        grouped_data = ticker_data.groupby(ticker_data.index).agg({
            'price': 'mean',  # Średnia cena w danej sekundzie
            'volume': 'sum'   # Suma volume w danej sekundzie
        })
        
        # Dodanie informacji o tickerze
        grouped_data['ticker'] = ticker
        
        # Dodanie kolumny z liczbą transakcji w danej sekundzie
        transaction_counts = ticker_data.groupby(ticker_data.index).size()
        grouped_data['transaction_count'] = transaction_counts
        
        # Zapisanie w słowniku
        ticker_dataframes[ticker] = grouped_data
        
        print(f"  - Przetworzono {len(ticker_data)} rekordów -> {len(grouped_data)} sekund")
        
        # Czyszczenie pamięci
        del ticker_data
        gc.collect()
    
    print("Przetwarzanie zakończone!")
    return ticker_dataframes

# Funkcja do wyświetlania statystyk
def show_ticker_stats(ticker_dataframes: Dict[str, pd.DataFrame]):
    """Wyświetla statystyki dla każdego tickera"""
    
    print("\n" + "="*60)
    print("STATYSTYKI TICKERÓW")
    print("="*60)
    
    total_original_records = 0
    total_processed_records = 0
    
    for ticker, df in ticker_dataframes.items():
        print(f"\nTicker: {ticker}")
        print(f"Liczba sekund: {len(df)}")
        print(f"Zakres dat: {df.index.min()} - {df.index.max()}")
        print(f"Średnia cena: {df['price'].mean():.6f}")
        print(f"Całkowity volume: {df['volume'].sum():.2f}")
        print(f"Całkowita liczba transakcji: {df['transaction_count'].sum()}")
        print(f"Średnio transakcji na sekundę: {df['transaction_count'].mean():.2f}")
        print(f"Użycie pamięci: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
        
        total_processed_records += len(df)
        total_original_records += df['transaction_count'].sum()
    
    print(f"\n" + "="*60)
    print(f"PODSUMOWANIE:")
    print(f"Oryginalne rekordy: {total_original_records}")
    print(f"Przetworzone sekundy: {total_processed_records}")
    print(f"Redukcja danych: {(1 - total_processed_records/total_original_records)*100:.2f}%")

# Funkcja do dostępu do konkretnego tickera
def get_ticker_data(ticker_dataframes: Dict[str, pd.DataFrame], ticker: str) -> pd.DataFrame:
    """Zwraca DataFrame dla konkretnego tickera"""
    if ticker in ticker_dataframes:
        return ticker_dataframes[ticker]
    else:
        print(f"Ticker {ticker} nie został znaleziony!")
        print(f"Dostępne tickery: {list(ticker_dataframes.keys())}")
        return None

# Funkcja do analizy konkretnego tickera
def analyze_ticker(ticker_dataframes: Dict[str, pd.DataFrame], ticker: str):
    """Przeprowadza podstawową analizę dla konkretnego tickera"""
    
    df = get_ticker_data(ticker_dataframes, ticker)
    if df is None:
        return
    
    print(f"\n" + "="*50)
    print(f"ANALIZA TICKERA: {ticker}")
    print("="*50)
    
    print(f"Pierwsze 5 rekordów:")
    print(df.head())
    
    print(f"\nTop 5 sekund z największym volume:")
    print(df.nlargest(5, 'volume')[['price', 'volume', 'transaction_count']])
    
    print(f"\nTop 5 sekund z największą liczbą transakcji:")
    print(df.nlargest(5, 'transaction_count')[['price', 'volume', 'transaction_count']])
    
    print(f"\nStatystyki opisowe:")
    print(df.describe())

# Funkcja do eksportu danych
def export_ticker_data(ticker_dataframes: Dict[str, pd.DataFrame], 
                      export_format: str = 'csv', 
                      path: str = './'):
    """
    Eksportuje dane do plików
    
    Parameters:
    ticker_dataframes: słownik z DataFrames
    export_format: 'csv' lub 'parquet'
    path: ścieżka do zapisu
    """
    
    import os
    if not os.path.exists(path):
        os.makedirs(path)
    
    for ticker, df in ticker_dataframes.items():
        if export_format == 'csv':
            filename = f"{path}{ticker}_data.csv"
            df.to_csv(filename)
        elif export_format == 'parquet':
            filename = f"{path}{ticker}_data.parquet"
            df.to_parquet(filename)
        
        print(f"Zapisano {ticker} do {filename}")

# Funkcja do szybkiego podglądu danych
def quick_preview(df: pd.DataFrame):
    """Szybki podgląd struktury danych"""
    print("STRUKTURA DANYCH:")
    print(f"Shape: {df.shape}")
    print(f"Columns: {list(df.columns)}")
    print(f"Index name: {df.index.name}")
    print(f"Index type: {type(df.index[0])}")
    print(f"Memory usage: {df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")
    
    print(f"\nPierwsze 3 rekordy:")
    print(df.head(3))
    
    print(f"\nUnikalne tickery: {df['ticker'].unique()}")
    print(f"Liczba unikalnych tickerów: {df['ticker'].nunique()}")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import mean_squared_error, mean_absolute_error
import warnings
warnings.filterwarnings('ignore')

def create_features_irregular(df):
    """
    Tworzy features dla nieregularnych danych transakcyjnych
    df powinien mieć kolumny: price, volume, ticker
    Index powinien być datetime (już skonwertowany z timestamp)
    """
    
    print(f"Tworzenie features dla {len(df)} transakcji...")
    
    # Sortowanie według czasu (na wszelki wypadek)
    df = df.sort_index()
    
    # Podstawowe informacje o czasie między transakcjami
    df['time_diff'] = df.index.to_series().diff().dt.total_seconds()
    df['time_diff_hours'] = df['time_diff'] / 3600
    
    # Returns oparte na czasie, nie na liczbie transakcji
    df['price_change'] = df['price'].pct_change()
    
    # Rolling windows oparte na czasie (nie na liczbie obserwacji)
    windows = ['1T', '5T', '15T', '1H', '4H', '24H']  # T = minuty, H = godziny
    
    for window in windows:
        window_name = window.replace('T', 'min').replace('H', 'h')
        
        # Price returns w oknie czasowym
        df[f'return_{window_name}'] = (
            df['price'].rolling(window).apply(lambda x: (x.iloc[-1] / x.iloc[0] - 1) if len(x) > 1 else np.nan)
        )
        
        # Volume w oknie czasowym
        df[f'volume_{window_name}'] = df['volume'].rolling(window).sum()
        
        # Liczba transakcji w oknie czasowym
        df[f'count_{window_name}'] = df['volume'].rolling(window).count()
        
        # Volatility (std returns) w oknie czasowym
        df[f'volatility_{window_name}'] = df['price_change'].rolling(window).std()
        
        # Min/Max price w oknie czasowym
        df[f'price_min_{window_name}'] = df['price'].rolling(window).min()
        df[f'price_max_{window_name}'] = df['price'].rolling(window).max()
        df[f'price_range_{window_name}'] = df[f'price_max_{window_name}'] - df[f'price_min_{window_name}']
    
    # Volume ratios (porównanie z dłuższymi okresami)
    df['volume_ratio_1h_24h'] = df['volume_1h'] / df['volume_24h']
    df['volume_ratio_current_1h'] = df['volume'] / df['volume_1h']
    
    # Trading intensity (transakcje na godzinę)
    df['trading_intensity_1h'] = df['count_1h'] / 1  # transakcje na godzinę
    df['trading_intensity_24h'] = df['count_24h'] / 24  # średnio transakcji na godzinę w ciągu 24h
    
    # Price momentum indicators
    df['price_momentum_1h'] = df['price'] / df['price'].rolling('1H').mean()
    df['price_momentum_24h'] = df['price'] / df['price'].rolling('24H').mean()
    
    # Czy to duża transakcja (powyżej 95 percentyla volume)
    df['large_transaction'] = (df['volume'] > df['volume'].quantile(0.95)).astype(int)
    
    # Czas od ostatniej dużej transakcji
    large_trans_times = df[df['large_transaction'] == 1].index
    df['time_since_large_trans'] = pd.NaT
    for idx in df.index:
        prev_large = large_trans_times[large_trans_times < idx]
        if len(prev_large) > 0:
            df.loc[idx, 'time_since_large_trans'] = (idx - prev_large[-1]).total_seconds() / 3600
    
    return df

def create_target_variable(df, target_horizon='1H'):
    """
    Tworzy zmienną target - zwrot w określonym horyzoncie czasowym
    """
    
    # Forward-looking return
    df['target_return'] = df['price'].pct_change(periods=1).shift(-1)  # Następna transakcja
    
    # Alternatywnie: return w określonym czasie w przyszłości
    df[f'target_return_{target_horizon}'] = (
        df['price'].rolling(target_horizon).apply(lambda x: (x.iloc[-1] / x.iloc[0] - 1) if len(x) > 1 else np.nan)
        .shift(-1)
    )
    
    return df

def resample_to_regular_intervals(df, interval='1H'):
    """
    Przekształca nieregularne dane na regularne interwały
    """
    
    print(f"Resampling do {interval} interwałów...")
    
    # Agregacja danych do regularnych interwałów
    resampled = df.resample(interval).agg({
        'price': 'last',  # Ostatnia cena w interwale
        'volume': 'sum',  # Suma volume
        'time_diff': 'mean',  # Średni czas między transakcjami
        'large_transaction': 'sum',  # Liczba dużych transakcji
        'trading_intensity_1h': 'last',
        'trading_intensity_24h': 'last',
        # Returns
        'return_1min': 'last',
        'return_5min': 'last', 
        'return_15min': 'last',
        'return_1h': 'last',
        'return_4h': 'last',
        'return_24h': 'last',
        # Volume features
        'volume_1h': 'last',
        'volume_24h': 'last',
        'volume_ratio_1h_24h': 'last',
        'volume_ratio_current_1h': 'last',
        # Volatility
        'volatility_1min': 'last',
        'volatility_5min': 'last',
        'volatility_15min': 'last',
        'volatility_1h': 'last',
        'volatility_4h': 'last',
        'volatility_24h': 'last',
        # Price features
        'price_momentum_1h': 'last',
        'price_momentum_24h': 'last',
        'price_range_1h': 'last',
        'price_range_24h': 'last',
        # Target
        'target_return': 'last',
        f'target_return_{interval}': 'last'
    })
    
    # Dodanie informacji o aktywności trading
    resampled['has_trading'] = (~resampled['price'].isna()).astype(int)
    resampled['transaction_count'] = df.resample(interval).size()
    
    return resampled

def prepare_ml_data(df, target_col='target_return'):
    """
    Przygotowuje dane do machine learning
    """
    
    # Feature columns (wykluczamy target i pomocnicze kolumny)
    exclude_cols = ['price', 'volume', 'time_diff', 'time_diff_hours', 'price_change', 
                   'target_return', 'target_return_1H', 'time_since_large_trans',
                   'price_min_1min', 'price_min_5min', 'price_min_15min', 'price_min_1h', 
                   'price_min_4h', 'price_min_24h', 'price_max_1min', 'price_max_5min',
                   'price_max_15min', 'price_max_1h', 'price_max_4h', 'price_max_24h']
    
    feature_cols = [col for col in df.columns if col not in exclude_cols]
    
    print(f"Wybrane features: {feature_cols}")
    
    # Przygotowanie danych
    X = df[feature_cols].copy()
    y = df[target_col].copy()
    
    # Usunięcie NaN
    mask = ~(X.isna().any(axis=1) | y.isna())
    X = X[mask]
    y = y[mask]
    
    print(f"Dane po usunięciu NaN: {len(X)} obserwacji")
    
    return X, y, feature_cols

def train_model_irregular_data(X, y, feature_cols):
    """
    Trenuje model na nieregularnych danych
    """
    
    print("Trenowanie modelu...")
    
    # Time series split
    tscv = TimeSeriesSplit(n_splits=3)
    
    # Model
    model = RandomForestRegressor(
        n_estimators=100, 
        random_state=42,
        max_depth=10,
        min_samples_split=5,
        min_samples_leaf=2
    )
    
    # Wyniki walidacji
    train_scores = []
    test_scores = []
    
    fold = 1
    for train_idx, test_idx in tscv.split(X):
        X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
        y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
        
        # Trenowanie
        model.fit(X_train, y_train)
        
        # Predykcje
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)
        
        # Metryki
        train_mse = mean_squared_error(y_train, train_pred)
        test_mse = mean_squared_error(y_test, test_pred)
        
        train_scores.append(train_mse)
        test_scores.append(test_mse)
        
        print(f"Fold {fold}: Train MSE = {train_mse:.6f}, Test MSE = {test_mse:.6f}")
        fold += 1
    
    # Finalne trenowanie na wszystkich danych
    model.fit(X, y)
    
    # Feature importance
    feature_importance = pd.DataFrame({
        'feature': feature_cols,
        'importance': model.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print(f"\nŚredni Train MSE: {np.mean(train_scores):.6f}")
    print(f"Średni Test MSE: {np.mean(test_scores):.6f}")
    
    print(f"\nTop 10 najważniejszych features:")
    print(feature_importance.head(10))
    
    return model, feature_importance

def process_ticker_with_ml(ticker_data, ticker_name):
    """
    Kompletny pipeline dla jednego tickera
    """
    
    print(f"\n" + "="*60)
    print(f"PRZETWARZANIE TICKERA: {ticker_name}")
    print("="*60)
    
    # Sprawdzenie czy mamy wystarczająco danych
    if len(ticker_data) < 100:
        print(f"Za mało danych dla {ticker_name}: {len(ticker_data)} transakcji")
        return None, None, None
    
    # Tworzenie features
    ticker_features = create_features_irregular(ticker_data)
    
    # Tworzenie target variable
    ticker_features = create_target_variable(ticker_features)
    
    # Resampling do regularnych interwałów (opcjonalnie)
    ticker_resampled = resample_to_regular_intervals(ticker_features, '1H')
    
    # Przygotowanie danych ML
    X, y, feature_cols = prepare_ml_data(ticker_resampled)
    
    if len(X) < 10:
        print(f"Za mało danych po przygotowaniu dla {ticker_name}: {len(X)} obserwacji")
        return None, None, None
    
    # Trenowanie modelu
    model, feature_importance = train_model_irregular_data(X, y, feature_cols)
    
    return model, feature_importance, ticker_resampled
```
